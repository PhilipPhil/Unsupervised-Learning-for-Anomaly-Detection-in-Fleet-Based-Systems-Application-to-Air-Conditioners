```python
import scipy.cluster.hierarchy as sch
from matplotlib.pyplot import figure

from random import seed
from random import random
from random import randint
seed(1)
```

```python
points = data.loc[np.r_[6:39, 46:61], :].iloc[:,:1495]
figure(figsize=(8, 6), dpi=80)
dendrogram = sch.dendrogram(sch.linkage(points, method='ward'))
```

```python
ALPHA = 0.9
for _ in range(20):
    state = []
    data = []
    indexs = []
    for _ in range(6):
        alpha = random()
        if alpha > ALPHA:
            state.append(1)
            index = randint(33, 47)
        else:
            state.append(0)
            index = randint(0, 32)
        indexs.append(index)
        data.append(points.iloc[index,:])
        
    if sum(state) == 0:
        data = pd.DataFrame(data)
        plt.figure(figsize=(8,4))
        plt.title(str(sum(state)) + ' : ' + str(state) + ' : ' + str(indexs), fontsize=12)
        dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
```
